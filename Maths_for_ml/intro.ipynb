```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 40, 1000) # start, finish, n points
```

#### Distance travelled by the robber is d = 2.5t

```python
d_robber = 2.5 * time
```

#### Distance travelled by sheriff: d=3(t-5)

```python
d_sherrif = 3 * (time - 5)
```

```python
# Create the figure and axes using plt.subplots()
fig, ax = plt.subplots()

plt.title("A Bank robber caught")
plt.xlabel("Time in minutes")
plt.ylabel("Distance in Kilometers")

# Set axis limits
ax.set_xlim(0,40)
ax.set_ylim(0, 100)

# Plot the data
ax.plot(time, d_robber, c="green", label="Bank robber")
ax.plot(time, d_sherrif, c="brown", label= "Sheriff")


ax.legend()
plt.axvline(x=30, color="purple", linestyle= "--")
plt.axhline(y= 75, color="purple", linestyle="--")
plt.show()
```
